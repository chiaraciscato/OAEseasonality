```python
dic_26 = xr.open_dataset('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/datasets/dic_ampl26.nc')
dic_70 = xr.open_dataset('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/datasets/dic_ampl70.nc')
```

```python
ph_26 = xr.open_dataset('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/datasets/ph_ampl26.nc')
ph_70 = xr.open_dataset('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/datasets/ph_ampl70.nc')
```

```python
oceanpco2_26 = xr.open_dataset('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/datasets/oceanpco2_ampl26.nc')
oceanpco2_70 = xr.open_dataset('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/datasets/oceanpco2_ampl70.nc')
```

```python
co2flux_26 = xr.open_dataset('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/datasets/co2flux_ampl26.nc')
co2flux_70 = xr.open_dataset('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/datasets/co2flux_ampl70.nc')
```

```python
cmapdiff = sns.diverging_palette(220, 20, as_cmap=True)
```

```python
fig, ((ax1, ax6), (ax2, ax7), (ax3, ax8), (ax4, ax9), (ax5, ax10)) = plt.subplots(5,2, subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(18, 32))
plt.subplots_adjust(wspace=0.005, hspace=0.06)

im1 = alk_26.alk.plot.pcolormesh( 
        ax=ax1, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-200,
        vmax=200,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im2 = ph_26.ph.plot.pcolormesh( 
        ax=ax2, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-0.1,
        vmax=0.1,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im3 = dic_26.dic.plot.pcolormesh( 
        ax=ax3, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-150,
        vmax=150,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im4 = oceanpco2_26.fco2.plot.pcolormesh( 
        ax=ax4, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-150,
        vmax=150,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im5 = co2flux_26.co2flux.plot.pcolormesh( 
        ax=ax5, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-10,
        vmax=10,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im6 = alk_26.alk.plot.pcolormesh( 
        ax=ax6, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-150,
        vmax=150,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im7 = ph_70.ph.plot.pcolormesh( 
        ax=ax7, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-0.1,
        vmax=0.1,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im8 = dic_70.dic.plot.pcolormesh( 
        ax=ax8, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-150,
        vmax=150,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im9 = oceanpco2_70.fco2.plot.pcolormesh( 
        ax=ax9, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-150,
        vmax=150,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im10 = co2flux_70.co2flux.plot.pcolormesh( 
        ax=ax10, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-10,
        vmax=10,
        cmap=cmapdiff, 
        add_colorbar=False
        )

axes = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10]

for ax in axes:
    ax.coastlines()
    ax.set_extent([-25, 10, 40, 70])
    ax.add_feature(cfeature.LAND, zorder=100, facecolor='lightgrey', edgecolor='black')
    ax.plot(3, 53, "o", color='black', markersize=3)
    ax.text(0.5, 54,'[52°N 3°E]', fontsize=12, color='black')

y = list(range(45,70))
x = list(range(-27,9))
    
for ax in axes: 
    if ax == ax1 or ax == ax2 or ax == ax3 or ax == ax4 or ax == ax5:
        ax.set_yticks(np.arange(min(y), max(y)+1, 5))
        labels_lat = [item.get_text() for item in ax.get_yticklabels()]
        lat_label = ['45°N', '50°N', '55°N', '60°N', '65°N']
        ax.set_yticklabels(lat_label, fontsize=12)
        ax.set_ylabel('')
    if ax == ax5 or ax == ax10:
        ax.set_xticks(np.arange(min(x)+7, max(x)+1, 5))
        labels_lon = [item.get_text() for item in ax.get_xticklabels()]
        lon_label = ['-20°E', '-15°E', '-10°E', '-5°E', '0°E', '5°E']
        ax.set_xticklabels(lon_label, fontsize=12)
        ax.set_xlabel('')
    if ax == ax1:
        ax.set_title("SSP1-2.6", fontsize=14, y=1.02)
    elif ax == ax6:
        ax.set_title("SSP3-7.0", fontsize=14, y=1.02),
    else:
        ax.set_title("")
        
for n, ax in enumerate(axes):
    ax.text(0.88, 0.9, string.punctuation[7] + ' ' + string.ascii_lowercase[n] + ' ' + string.punctuation[8],
            transform=ax.transAxes, size=12, weight='bold', color='black')

cbar1 = fig.colorbar(im1, ax = [ax1, ax6], aspect = 40, pad=0.02, location='right')
cbar2 = fig.colorbar(im2, ax = [ax2, ax7], aspect = 40, pad=0.02, location='right', ticks=[-0.1, -0.05, 0, 0.05, 0.1])
cbar3 = fig.colorbar(im3, ax = [ax3, ax8], aspect = 40, pad=0.02, location='right')
cbar4 = fig.colorbar(im4, ax = [ax4, ax9], aspect = 40, pad=0.02, location='right')
cbar5 = fig.colorbar(im5, ax = [ax5, ax10], aspect = 40, pad=0.02, location='right', ticks=[-10, -5, 0, 5, 10])

cbars = [cbar1, cbar2, cbar3, cbar4, cbar5]
    
for cbar in cbars:
    if cbar == cbar1:
        cbar.set_label('Δ Alkalinity [mmol/$\mathregular{m^3}$]', fontsize=14)
        cbar.ax.tick_params(labelsize=12)
    elif cbar == cbar2:
        cbar.set_label('Δ pH', fontsize=14)
        cbar.ax.tick_params(labelsize=12)
        cbar.ax.set_yticklabels(['-0.1', '-0.05', '0', '0.05', '0.1'])
    elif cbar == cbar3:
        cbar.set_label('Δ DIC [mmol/$\mathregular{m^3}$]', fontsize=14)
        cbar.ax.tick_params(labelsize=12)
    elif cbar == cbar4:
        cbar.set_label('Δ Ocean p$\mathregular{CO_{2}}$ [µatm]', fontsize=14)
        cbar.ax.yaxis.get_offset_text().set(size=14)
        cbar.ax.tick_params(labelsize=12)
    else:
        cbar.set_label('Δ $\mathregular{CO_{2}}$ flux $\mathregular{[mol/m^2/yr]}$', fontsize=14)
        cbar.ax.yaxis.get_offset_text().set(size=14)
        cbar.ax.tick_params(labelsize=12)
        cbar.ax.set_yticklabels(['-10', '-5', '0', '5', '10'])

pass

plt.savefig('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/seasonality/figures/amplitude.png')
```
