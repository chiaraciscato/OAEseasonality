```python
class HandlerEllipse(HandlerPatch):
    def create_artists(self, legend, orig_handle,
                       xdescent, ydescent, width, height, fontsize, trans):
        center = 0.5 * width - 0.5 * xdescent, 0.5 * height - 0.5 * ydescent
        p = mpatches.Ellipse(xy=center, width=height + xdescent,
                             height=height + ydescent)
        self.update_prop(p, orig_handle, legend)
        p.set_transform(trans)
        return [p]
```

```python
def figpco2(base, oae, delta, axTop, axBot):
    axTop.plot(base.fco2, '-', color="black", linewidth=0.5, zorder=1)
    axTop.plot(oae.fco2, '-', color="black", linewidth=0.5, zorder=1)
    axTop.scatter(base.time_counter, base.fco2, color=base.time_counter.map(lambda x: colors_month_fco2[x]), edgecolors='black',  linewidth=0.5, s=70)
    axTop.scatter(oae.time_counter, oae.fco2, color=oae.time_counter.map(lambda x: colors_month_fco2[x]), edgecolors='black',  linewidth=0.5, s=70)
    axBot.plot(delta.fco2, '-', color="black", linewidth=0.5, zorder=1)
    axBot.scatter(delta.time_counter, delta.fco2, color=delta.time_counter.map(lambda x: colors_month_fco2[x]), edgecolors='black',  linewidth=0.5, s=70)
    axTop.tick_params(axis='y', which='major', labelsize=12)
```
