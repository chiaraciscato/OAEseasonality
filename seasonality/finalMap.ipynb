```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2,3, subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(23, 16))
plt.subplots_adjust(wspace=0.07, hspace=0.05)

alk_26.alk.plot.pcolormesh( 
        ax=ax1, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-200,
        vmax=200,
        cmap=cmapdiff, 
        add_colorbar=False
        )

oceanpco2_26.fco2.plot.pcolormesh( 
        ax=ax2, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-150,
        vmax=150,
        cmap=cmapdiff, 
        add_colorbar=False
        )

co2flux_26.co2flux.plot.pcolormesh( 
        ax=ax3, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-10,
        vmax=10,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im1 = alk_26.alk.plot.pcolormesh( 
        ax=ax4, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-150,
        vmax=150,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im2 = oceanpco2_70.fco2.plot.pcolormesh( 
        ax=ax5, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-150,
        vmax=150,
        cmap=cmapdiff, 
        add_colorbar=False
        )

im3 = co2flux_70.co2flux.plot.pcolormesh( 
        ax=ax6, 
        x="nav_lon", 
        y="nav_lat",
        vmin=-10,
        vmax=10,
        cmap=cmapdiff, 
        add_colorbar=False
        )

axes = [ax1, ax2, ax3, ax4, ax5, ax6]

for ax in axes:
    ax.coastlines()
    ax.set_extent([-25, 10, 40, 70])
    ax.add_feature(cfeature.LAND, zorder=100, facecolor='lightgrey', edgecolor='black')
    ax.plot(3, 53, "o", color='black', markersize=3)
    ax.text(0.5, 54,'[52°N 3°E]', fontsize=12, color='black')

y = list(range(45,70))
x = list(range(-27,9))
    
for ax in axes: 
    if ax == ax1 or ax == ax4:
        ax.set_yticks(np.arange(min(y), max(y)+1, 5))
        labels_lat = [item.get_text() for item in ax.get_yticklabels()]
        lat_label = ['45°N', '50°N', '55°N', '60°N', '65°N']
        ax.set_yticklabels(lat_label, fontsize=14)
        ax.set_ylabel("")
    if ax == ax4 or ax == ax5 or ax == ax6:
        ax.set_xticks(np.arange(min(x)+7, max(x)+1, 5))
        labels_lon = [item.get_text() for item in ax.get_xticklabels()]
        lon_label = ['-20°E', '-15°E', '-10°E', '-5°E', '0°E', '5°E']
        ax.set_xticklabels(lon_label, fontsize=14)
        ax.set_xlabel('')
    if ax == ax1:
        ax.set_title("Δ Alkalinity [mmol/$\mathregular{m^3}$]", fontsize=16, y=1.16)
    elif ax == ax2:
        ax.set_title("Δ Ocean p$\mathregular{CO_{2}}$ [µatm]", fontsize=16, y=1.16),
    elif ax == ax3:
        ax.set_title("Δ $\mathregular{CO_{2}}$ flux $\mathregular{[mol/m^2/yr]}$", fontsize=16, y=1.16),
    else:
        ax.set_title("")
        
for n, ax in enumerate(axes):
    ax.text(0.88, 0.9, string.punctuation[7] + string.ascii_lowercase[n] + string.punctuation[8],
            transform=ax.transAxes, size=18, weight='bold', color='black')

cbar1 = fig.colorbar(im1, ax = [ax1, ax4], aspect = 40, pad=0.02, location='top')
cbar2 = fig.colorbar(im2, ax = [ax2, ax5], aspect = 40, pad=0.02, location='top')
cbar3 = fig.colorbar(im3, ax = [ax3, ax6], aspect = 40, pad=0.02, location='top', ticks=[-10, -5, 0, 5, 10])

cbars = [cbar1, cbar2, cbar3]
    
for cbar in cbars:
    if cbar == cbar1:
        cbar.ax.tick_params(labelsize=12)
    elif cbar == cbar2:
        cbar.ax.yaxis.get_offset_text().set(size=14)
        cbar.ax.tick_params(labelsize=12)
    else:
        cbar.ax.yaxis.get_offset_text().set(size=14)
        cbar.ax.tick_params(labelsize=12)
        cbar.ax.set_xticklabels(['-10', '-5', '0', '5', '10'])

pass

plt.savefig('/Users/chiaraciscato/Desktop/geomar/OAEseasonality/seasonality/figures/map.png')
```
